```python
import pandas as pd
path = '/content/drive/MyDrive/CSV/tahun-20172018-ganjil---data-jumlah-sekolah-dasar-di-kecamatan-andir.csv'
df = pd.read_csv(path)
df.describe()
```

```python
df
```

```python
df.info
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Rombongan Belajar', y='Guru')
plt.title('indeks kesejahtraan warga')
plt.xlabel('jumlah penduduk')
plt.ylabel('indeks')
plt.xlim(0, 50)
plt.ylim(0, 70)
plt.grid(True)
plt.show()
```

```python
import numpy as np
X = np.array(df['Rombongan Belajar'])
y = np.array(df['Guru'])
print(f'X : {X}')
print(f'y : {y}')
```

```python
X = X.reshape(-1, 1)
X.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
X_vis = np.array([0, 10000]).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis, 'r')
plt.title('indeks kesejahtraan warga')
plt.xlabel('jumlah penduduk')
plt.ylabel('indeks')
plt.xlim(0, 50)
plt.ylim(0, 70)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ekskul = df[ 'Rombongan Belajar']
mengikuti = df['Guru']
c = 'black', 'pink', 'blue'
plt.title("Lingkungan")
plt.bar(ekskul, mengikuti, color = c)
plt.show()
```

```python
print("")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("usia")
ekskul = df[ 'Rombongan Belajar']
mengikuti = df['Guru']
plt.title("Lingkunagn")
ax.pie(mengikuti,labels=ekskul,autopct='%1.2f%%')
plt.show()
```

```python
Guru = np.array([2330, 5200, 8800, 3200, 1200]).reshape(-1, 1)
Guru
```

```python
prediksi_indeks = model.predict(Guru)
prediksi_indeks
```

```python
for Guru, index in zip(Guru, prediksi_indeks):
  print(f'Jumlah Penduduk : {Guru} prediksi indeks : {index}')
```
